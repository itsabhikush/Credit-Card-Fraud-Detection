## Dataset Information

The dataset contains transactions made by credit cards in September 2013 by European cardholders. This dataset presents transactions that occurred in two days, where we have 492 frauds out of 284,807 transactions. The dataset is highly unbalanced, the positive class (frauds) account for 0.172% of all transactions.


The dataset is characterized by the absence of original features and additional contextual information due to confidentiality constraints. Principal Component Analysis (PCA) has been applied to derive numerical input variables, denoted as V1 through V28. However, two features, 'Time' and 'Amount,' have not undergone PCA transformation. The 'Time' feature signifies the elapsed seconds between each transaction and the initial transaction in the dataset, while the 'Amount' feature denotes the transaction amount. The latter feature can be particularly relevant for methodologies involving example-dependent cost-sensitive learning.

The response variable is represented by the 'Class' feature, taking a value of 1 in cases of fraud and 0 otherwise.

Given the substantial class imbalance, it is recommended to assess accuracy using the Area Under the Precision-Recall Curve (AUPRC). Traditional measures such as the confusion matrix accuracy may not provide meaningful insights in the context of unbalanced classification.

## Input Split

```python
X = df.drop(columns=['Class'], axis=1)
y = df['Class']
```

## Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaler = sc.fit_transform(X)
```

```python
x_scaler[-1]
```

## Model Training

```python
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)
# training
model.fit(x_train, y_train)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

## Class Imbalancement

```python
sns.countplot(y_train)
```

```python

# balance the class with equal distribution
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
sns.countplot(y_smote)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
# training 
model.fit(x_smote, y_smote)
# testing 
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=-1)
# training
model.fit(x_smote, y_smote)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_jobs=-1)
# training
model.fit(x_smote, y_smote)
# testing
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```
